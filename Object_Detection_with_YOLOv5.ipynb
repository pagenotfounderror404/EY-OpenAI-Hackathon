```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#Uswing the small model due to computation contraint you can check out "https://pytorch.org/hub/ultralytics_yolov5/" for larger model which might give better results
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
img = 'image.jpeg'
```

```python
results = model(img)
results.print()
```

```python
plt.imshow(np.squeeze(results.render()[0]))
plt.show()
```

```python
results.render()[0].shape
```

```python
cap = cv2.VideoCapture(0) #You can also video directory here
while cap.isOpened():
    ret, frame = cap.read()
    
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
